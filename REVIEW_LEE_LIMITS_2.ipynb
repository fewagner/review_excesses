```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [10, 6.5]

import darklim
```

```python
def get_spectra_exp(e0, tot_rate_hz, exposure_time, res=0):
    total_events = tot_rate_hz * exposure_time
    return np.random.exponential(size=int(total_events))*e0 + np.random.normal(size=int(total_events))*res

    
```

```python
resolution = 1 #eV
threshold = 5*resolution #eV
energy_range = np.arange(threshold, 100, 0.5) #eV
exposure_s = 60 * 1
efficiencies = np.ones_like(energy_range)

dm_masses = np.geomspace(0.05, 5, num=25)

res_arr = [0.1, 0.2, 0.3, 0.5, 1, 2, 3, 5]

spectra_arr = []
i = 0
while i < len(res_arr):
    spectra_arr.append(get_spectra_exp(20, 10, exposure_s))
    i += 1
    
    
limits_arr = []

i = 0
while i < len(res_arr):
    sig, ol1, ol2 = darklim.limit.optimuminterval(spectra_arr[i]*1e-3, energy_range*1e-3, efficiencies,
                                                  dm_masses, exposure_s/(24*60*60), res=res_arr[i]*1e-3,
                                                 verbose=True, hard_threshold=res_arr[i]*5e-3)
    limits_arr.append(sig)
    print("Done with spectrum " + str(i) + " of " + str(len(spectra_arr)))
    print(" ")
    i += 1
```

```python
i = 0
while i < len(limits_arr):
    plt.plot(dm_masses, limits_arr[i], label = "Resolution: " + str(res_arr[i]) + " eV")
    i += 1
    
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
